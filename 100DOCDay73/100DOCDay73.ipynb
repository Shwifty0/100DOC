```python
reshaped_df = df.pivot(index="DATE", columns="TAG", values="POSTS")
reshaped_df.head()
```

```python
reshaped_df.fillna(value=0, inplace=True)
reshaped_df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
for i in range(len(reshaped_df.columns)):
    p_lang = reshaped_df.columns[i]
    plt.plot(reshaped_df[p_lang].rolling(window=5).mean(), label = p_lang, linewidth = 0.8) 
    plt.xlabel("Year"), plt.ylabel("Posts")

plt.legend(fontsize = 16)
```
